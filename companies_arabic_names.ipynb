# Replace the numeral labels with Arabic names

```python
import pandas as pd

data_tags = pd.read_csv('Data.csv')
new_data = [int(x.split('.')[0]) for x in data_tags.iloc[0, 1:]]
data_tags.iloc[0, 1:] = new_data
data_names = pd.read_csv('Tags_with_names.csv')
# Create a dictionary for mapping symbols to their names
mapping_dict = dict(zip(data_tags.iloc[0, 1:], data_names.iloc[:,1]))

# Replace symbols in the second row with their names using the mapping dictionary
data_tags.iloc[0] = data_tags.iloc[0].replace(mapping_dict)

# Display the first few rows of the modified dataframe
data_tags.to_csv('Named_TASi_Companies.csv', index=False)
```

```python
import pandas as pd

# Read the uploaded file with names
data_with_names_df = pd.read_csv('Named_TASi_Companies.csv')

# Extract unique company names from the second row
company_names = data_with_names_df.iloc[0].unique()

# Remove NaN values (if any) from the company names list
company_names = [name for name in company_names if pd.notna(name)]

# Retry saving separate CSV files for each company
saved_files_retry = []

for company in company_names:
    # Extract columns related to the current company
    company_data = data_with_names_df[['Date'] + [col for col in data_with_names_df.columns if company == data_with_names_df[col][0]]]
    # Remove the row with company names to keep only data
    company_data = company_data.drop(0).reset_index(drop=True)
    # Save to CSV
    filename = f"Data/{company}.csv"
    company_data.to_csv(filename, index=False)
    saved_files_retry.append(filename)

saved_files_retry
```
