# Replace the numeral labels with Arabic names

```python
import pandas as pd

data_tags = pd.read_csv('Data.csv')
new_data = [int(x.split('.')[0]) for x in data_tags.iloc[0, 1:]]
data_tags.iloc[0, 1:] = new_data
data_names = pd.read_csv('Tags_with_names.csv')
# Create a dictionary for mapping symbols to their names
mapping_dict = dict(zip(data_tags.iloc[0, 1:], data_names.iloc[:,1]))

# Replace symbols in the second row with their names using the mapping dictionary
data_tags.iloc[0] = data_tags.iloc[0].replace(mapping_dict)

# Display the first few rows of the modified dataframe
data_tags.to_csv('Named_TASi_Companies.csv', index=False)
```

```python
import pandas as pd

# Read the uploaded file with names
data_with_names_df = pd.read_csv('Named_TASi_Companies.csv')

# Extract unique company names from the second row
company_names = data_with_names_df.iloc[0].unique()

# Remove NaN values (if any) from the company names list
company_names = [name for name in company_names if pd.notna(name)]

# Retry saving separate CSV files for each company
saved_files_retry = []

for company in company_names:
    # Extract columns related to the current company
    company_data = data_with_names_df[['Date'] + [col for col in data_with_names_df.columns if company == data_with_names_df[col][0]]]
    # Remove the row with company names to keep only data
    company_data = company_data.drop(0).reset_index(drop=True)
    # Save to CSV
    filename = f"Data/{company}.csv"
    company_data.to_csv(filename, index=False)
    saved_files_retry.append(filename)

saved_files_retry
```

## Combine all data information into one file

```python
import pandas as pd
import os

# 1. Load the necessary datasets
companies_df = pd.read_csv("./Data/Companies_with_names.csv")
sp500_df = pd.read_csv("./Data/SP500.csv")
sp500_df['Date'] = pd.to_datetime(sp500_df['Date'])
tasi_df = pd.read_csv("./Data/TASi.csv")
tasi_df['Date'] = pd.to_datetime(tasi_df['Date'])
repo_oil_df = pd.read_csv("./Data/REPO_Oil_Inf_GDP_Unemp.csv")
repo_oil_df['Date'] = pd.to_datetime(repo_oil_df['Date'])

# Extrapolate the monthly data to daily data for the REPO_Oil_Inf_GDP_Unemp file
daily_data_list = []
for _, row in repo_oil_df.iterrows():
    month_start = row['Date']
    month_end = pd.Timestamp(month_start.year, month_start.month, 1) + pd.offsets.MonthEnd(0)
    date_range = pd.date_range(month_start, month_end)
    repeated_data = pd.concat([row.to_frame().T] * len(date_range))
    repeated_data['Date'] = date_range
    daily_data_list.append(repeated_data)
daily_repo_oil_df = pd.concat(daily_data_list, ignore_index=True)

# Load the reference company file as a guide for the date range
reference_company_df = pd.read_csv("./Data/TASi Companies Data/البنك الأهلي السعودي.csv")
reference_company_df['Date'] = pd.to_datetime(reference_company_df['Date'])
reference_date_range = reference_company_df['Date']

# Fetch the company datasets
all_company_data = []
for _, row in companies_df.iterrows():
    company_name_arabic = row['أسم الشركة']
    company_code = row['رمز الشركة']
    company_sector = row[' القطاع']
    file_path = f"./Data/TASi Companies Data/{company_name_arabic}.csv"
    if os.path.exists(file_path):
        company_data_df = pd.read_csv(file_path)
        company_data_df['Date'] = pd.to_datetime(company_data_df['Date'])
        company_data_df['code'] = company_code
        company_data_df['sector'] = company_sector
        all_company_data.append(company_data_df)

# Begin the merging process using the reference date range
merged_data_based_on_reference = []
for df in all_company_data:
    df_filtered = df[df['Date'].isin(reference_date_range)]
    merged_df = pd.merge(df_filtered, daily_repo_oil_df, on='Date', how='left')
    merged_df = pd.merge(merged_df, sp500_df, on='Date', how='left', suffixes=('', '_SP500'))
    merged_df = pd.merge(merged_df, tasi_df, on='Date', how='left', suffixes=('', '_TASi'))
    merged_data_based_on_reference.append(merged_df)

# Concatenate and save the final data
final_data_based_on_reference = pd.concat(merged_data_based_on_reference, ignore_index=True)
final_data_based_on_reference.to_csv("./Data/Mega_Companies_Info.csv", index=False)
```
